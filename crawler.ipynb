```python
while True:
    record = col.find({'crawled':False}).skip(random.randint(0,1000)).next()
    driver.get(record['url'])
    if driver.current_url == 'https://www.linkedin.com/404/':
        col.delete_one({'_id': record['_id']})
        continue
    # Crawl related user
    related_users = crawl_relation_url(driver=driver, url=record['url'])
    # Crawl profile
    profile = crawl_profile(driver=driver, url=record['url'])
    # Insert database
    ## Update profile
    col.find_one_and_update({'_id': record['_id']}, {'$set': profile})
    ## Insert related users
    if type(related_users) == list:
        for user in related_users:
            if is_vietnamese(user['fullname']) and col.find_one({'url': user['url']}) == None:
                col.insert_one(user)
    time.sleep(random.randint(2,3))
    break
```
